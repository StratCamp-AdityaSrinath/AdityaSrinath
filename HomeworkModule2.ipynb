```python
# 1: Tuples

# Part 1

# Tuples are a type of data structure that are immutable and ordered, but which allow duplicates

# Part 2

# Define the list with nested tuples
tupelist = [(1, 4, 7), (2, 5, 8), (3, 6, 9)]

# Use tup to instruct python to access elements of a tuple, and sum them. Then print the highest value.
print(max(sum(tup) for tup in tupelist))
```

```python
# 2: Sets

# Part 1

# Sets are a type of data structure that have elements that are mutable and unordered, but which do not allow duplicates

# Part 2

lst = [1, 2, 3, 3, 4, 1]
st1 = set(lst)

print(st1)

# Part 3

st2 = {1, 2, 3}
st3 = {2, 4, 6}
st4 = {2, 5, 8}

print(st2&st3&st4)
```

```python
# 3: Dictionaries

# Part 1

# Dictionaries are a type of data structure that have elements that are mutable, with key-value pairs that have unique mappings.

# Part 2

dict1 = {"TC" : 34, "DK" : 32, "FB" : 48}
dict2 = {"AP" : 33, "DR" : 45, "DT" : 30}

# Merge by "updating" dict1 with contents of dict2
dict1.update(dict2)

print(dict1)

# Merge by using the double * to "unpack" and merge the key-value pairs of two dictionaries
merged_dict = { **dict1, **dict2}

print(merged_dict)

# Part 3

dict3 = {"First": 545, "Second": 2, "Third": 122}

# Delete a specified key from the dictionary, deleting the value as a result
del dict3["Second"]

print(dict3)

# Add a key-value pair by "updating" the original dictionary with the new key
dict3.update({"Fourth": 175})

print(dict3)
```

```python
# 4: Strings

# Part 1

print("Enter a name")

s1 = str(input())

# Turn everything to lowercase to ease with IDing from list
s2 = s1.lower()

vowels = ["a", "e", "i", "o", "u"]

s3 = "" # Use empty strings to create new words with replaced characters

for i in range(len(s2)): # Use len to get around python restrictions on strings
    if s2[i] in vowels:
        s3 = s3 + s2[i].replace(s2[i], "*") # Replace vowels with *
    else:
        s3 = s3 + s2[i] # Leave consonants alone
print(s3)

# Part 2

print("Enter a word in English")

s4 = str(input())

print(s4[::-1])

# Part 3

print("Is this what we're looking for?")

s5 = input()
at_sign = s5.count("@")
banned = ["!", "#", "$", "%", "^", "&", "*", "(", ")", "+", "=", "|", "\\", "}", "]", "[", "{", "'", ":", ";", ">", "<", "?", "/", "~", "`"]
at_position = s5.find("@")
dot_count = s5.count(".")
dot_first_position = s5.find(".")
dot_last_position = s5.rfind(".")
smart_ass_check = int(at_position) - int(dot_position)
super_smart_ass_check = int(s5.find(".", at_position)) - int(at_position)

if any(s5) != any(banned) and len(s5) >= 6 and at_sign == 1 and at_position != 0 and dot_count >= 1 and dot_first_position != 0 and s5[-1] != "." and ".." not in s5 and smart_ass_check != 1 and super_smart_ass_check != 1:
    print("True")
else:
    print("False")

# Part 4

while True:
    try:
        user_name = str(input("What is your name? "))
    except ValueError:
        print("Invalid input. Please provide your name only.")
    else:
        break

while True:
    try:
        user_age = int(input("What is your age? "))
    except ValueError:
        print("Invalid input. Please provide integers only.")
    else:
        break

while True:
    try:
        user_city = str(input("Which city do you live in? "))
    except ValueError:
        print("Invalid input. Please provide your city only.")
    else:
        break

while True:
    try:
        user_phone = int(input("What is your phone number? "))
    except ValueError:
        print("Invalid input. Please provide integers only.")
    else:
        break

print(f"Your name is {user_name}, you are {user_age} years old, you live in {user_city}, and your phone number is {user_phone}.")
```
